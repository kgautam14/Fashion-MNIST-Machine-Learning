```python
import pandas
from sklearn.model_selection import KFold
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVR
import numpy as np
from sklearn.model_selection import cross_val_score

dataTrain = pandas.read_csv('modDataTrain.csv')
dataTest = pandas.read_csv('modDataTest.csv')
```

```python
yTrain = dataTrain.iloc[:,0]
yTest = dataTest.iloc[:,0]

xTrain = dataTrain.drop(['label'], axis = 1)
xTest = dataTest.drop(['label'], axis = 1)
```

```python
display(xTest)
```

```python
scl = MinMaxScaler(feature_range = (0,1))
X_Train = scl.fit_transform(xTrain)
Y_Train = yTrain
X_Test = scl.fit_transform(xTest)
Y_Test = yTest
```

```python
display(Y_Test)
```

```python
 
best_svr = SVR(kernel = 'rbf')

CV = KFold(n_splits = 5, random_state = 123, shuffle = True)

# training data with SVR
best_svr.fit(X_Train, Y_Train)

scr = cross_val_score(best_svr, X_Train, Y_Train, cv = CV)


display(scr)
```

```python
from sklearn.linear_model import LogisticRegression
import sklearn

X_Train = sklearn.preprocessing.scale(X_Train)

lr = LogisticRegression()
lr.fit(X_Train, Y_Train)
cv_scores = cross_val_score(lr, X_Train, Y_Train, cv = CV)

display(cv_scores)
```

```python
display(scr.mean())
display(cv_scores.mean())
```
